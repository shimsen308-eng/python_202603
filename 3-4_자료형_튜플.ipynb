## 2.튜플의 요소값을 지우거나 변경하면?

```python
t1 = (1,2,'a','b')
del t1[0]
#튜플은 del 안됨
```

```python
t1[0] = 'c'
#변경 불가
```

## 3.튜플 다루기

```python
# 인덱싱
t1 = (1,2,'a','b')
t1[0]
```

```python
t1[-1]
```

```python
# 슬라이싱
t1[:2]
```

```python
t1[2:]
```

```python
# 튜플 더하기 : 연결
t1 = (1,2,'a','b')
t2 = (3,4)

t3 = t1 + t2
t3
```

```python
# 튜플 곱하기 :  반복
t3 = t2*3
t3
```

```python
# 튜플 길이 :  Len()
len(t1)
```

```python
len(t2)
```

```python
len(t3)
```
